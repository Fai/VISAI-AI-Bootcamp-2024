### Lab 3

```python
# Example 2

# normal distribution
def normal(x, mean , sd):
    pdf_value = norm.pdf(x, mean, sd)
    print("prob_density of x = ", pdf_value)
    
    
    # draw graph
    plt.axvline(x=x, color='red')
    x_linespace = np.linspace(mean - 4*sd, mean + 4*sd, 200)
    pdf = norm.pdf(x_linespace, mean, sd)
    
    #  Plotting the Results 
    plt.plot(x_linespace, pdf)
    plt.xlabel('Data points')
    plt.ylabel('Probability Density')
    return 

# standard normal distribution
def standard_normal(x, mean, sd):    
    x_standard = (x-mean)/sd
    normal(x_standard, 0 , 1)
    print('z = ', x_standard)
    return

# Data A
x1=38
mean1=26
sd1 = 12

# Data B
x2=32
mean2=17
sd2 = 15

# normal(x1, mean1, sd1)
# normal(x2, mean2, sd2)
# standard_normal(x1, mean1, sd1)
standard_normal(x2, mean2, sd2)

 
```

# Law of Large Number

### Lab 4

```python
# Example 4

def normal():  
    return 0, np.random.normal(0, 1)

def uniform():
    return 50, np.random.uniform(low=0, high=100)
    
def beta():
    return 2/7, np.random.beta(2, 5)

def bernoulli():
    choice = np.array([0, 0, 0, 1])
    return 1/4, np.random.choice(choice)

def law_of_large_number(n, distribution):
    sample =[]
    sample_mean=[]
    
    # random sample size i to find sample_mean
    for i in range(0, n):
        sample += [distribution()[1]]
        sample_mean += [np.mean(sample)]
        
    plt.xlim(-5,  n+5)
    plt.plot(list(range(1, n+1)), sample_mean, zorder=0)
    plt.hlines(distribution()[0], xmin=0, xmax=n, zorder=1)    

# n = Number of max samples
n= 1000

# choose distribution
# distribution = normal
# distribution = uniform
# distribution = beta
distribution = bernoulli

# number of trials
t_trials=4

for i in range(0, t_trials):
    law_of_large_number(n, distribution)
```

# Central limit Theorem

### Lab 5

```python
#Example 6

def normal(size):    
    global xlim_min
    global xlim_max    
    xlim_min=-2 
    xlim_max=2
    return np.random.normal(0, 1, size=size)

def uniform(size):
    global xlim_min
    global xlim_max 
    xlim_min=0    
    xlim_max=100
    return np.random.uniform(low=0, high=100, size=size)
    
def beta(size):
    global xlim_min
    global xlim_max 
    xlim_min=0    
    xlim_max=1
    return np.random.beta(2, 5, size=size)

def bernoulli(size):
    global xlim_min
    global xlim_max 
    choice = [0, 0, 0, 1]
    xlim_min=0    
    xlim_max=1
    return np.random.choice(choice, size=size)

def central_limit_thm(t, n, distribution):
    sample_mean=[]
    
    # random sample size n to find sample_mean, then do this t_points times and then find sample_mean_mean
    for i in range(0, t):
        sample = distribution(n)
        sample_mean += [np.mean(sample)]

    # print(sample_mean)
    sample_mean_mean = np.mean(sample_mean)
    sample_mean_variance = np.var(sample_mean)
    print('Mean = ', sample_mean_mean, 'Variance = ', sample_mean_variance)    
    plt.xlim(xlim_min,  xlim_max)
    plt.hist(sample_mean, bins=n_bins)
    return

# number of bins
n_bins = 20
# number of sample_mean
t_points = 1000

# choose distribution
# distribution = normal
# distribution = uniform
# distribution = beta
distribution = bernoulli

# n = Number of samples for calculating a sample_mean
for n in [100, 500,1000]:
    central_limit_thm(t_points, n, distribution)
```

# Covariance and Correlation

### Lab 6

```python
#Example 7

data_1 = np.array([12, 16, 28, 37, 35, 43])
data_2 = np.array([135, 167, 156, 168, 180, 200])

mat = np.stack((data_1, data_2)) 
cov_mat = np.cov(mat)
cor_mat = np.corrcoef(mat)
print("Covariance matrix = \n", cov_mat)
print("Correlation matrix = \n", cor_mat)

for i in range(len(mat)):
    plt.scatter(list(range(1,len(mat[i])+1)), mat[i])
    
plt.legend(("data 1","data 2"), loc='upper left')
```
