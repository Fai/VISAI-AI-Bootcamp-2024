### Mean, Median, Mod, and Percentile

$$mean=\dfrac{\sum x}{n}, \qquad median=ùë•_{(ùëõ+1)/2}$$

```python
data_set = np.array([3,7,5,8,4,3,2,4,9,3])
```

```python
# find mean and median of data_set
mean_data = np.mean(data_set)
median_data = np.median(data_set)

print("mean: ", mean_data)
print("median: ", median_data)
```

```python
# mod of data_set
values, counts = np.unique(data_set, return_counts=True)
print(values)
print(counts)
i  = np.argmax(counts)
print("argmax: ", i) 
print("mode: ", values[i])
```

$$ Standard\,Deviation\,( \sigma ) = \sqrt{\dfrac{\sum(x-\mu)^2}{n}}$$

```python
# find range, std, and variance of data_set
range_data = np.max(data_set) - np.min(data_set)
std_data = np.std(data_set)
variance_data = np.var(data_set)

print("range: ", range_data)
print("std: ", std_data)
print("variance: ", variance_data)
```

$$Inter\, Quartile\, Range\, (IQR)=ùëÑ_3‚àíùëÑ_1=ùë•_{3(ùëõ+1)/4}‚àíùë•_{(ùëõ+1)/4}$$

```python
# find inter quartile range
q1 = np.quantile(data_set, .25)
q2 = np.quantile(data_set, .50)
q3 = np.quantile(data_set, .75)

print(q1, q2, q3)
print("IQR: ", q3-q1)
```

# Law of Large Number and Central Limit Theorem

$X$ has $E[X]=\mu, Var[X]=\sigma^2$. <br> <br>
Law of Large Number: $\qquad  ùëõ‚Üí‚àû, \,\,     \overline{X}_n    \xrightarrow[]{}  \mu$ <br> <br>
Central Limit Theorem:  $\qquad  ùëõ‚Üí‚àû, \,\,   \overline{X}_n    \xrightarrow[]{d}  N\biggl(\mu,\dfrac{\sigma^2}{n}\biggr) \qquad$ where $E[\overline{X}]=\mu, Var[\overline{X}]=\dfrac{\sigma^2}{n}$

```python
n=1
prob=[]
flip=[]

n_filp = 10 # change this number

while n <= n_filp:
    head=0
    tail=0

    for i in range(n):
        if np.random.randint(2, size=1)[0] ==0:
            head+=1
        else:
            tail+=1
    k=head/(head+tail)
    prob.append(k)
    flip.append(n)
    n+=1

plt.plot(flip,prob)
plt.xlabel('Number of Flips')
plt.ylabel('Probability of Heads')
plt.grid(True)

plt.show()
```

# Covariance and Correlation

$$\text{Cov}(ùëã,ùëå)=E[(ùëã‚àí\mu_ùëã )(ùëå‚àí\mu_ùëå )]$$
Cov$(ùëã,ùëå)>0$:	large values of¬†$ùëã$ tend to happen with large values of¬†$ùëå$. <br>
Cov$(ùëã,ùëå)<0$:	$ùëã$ tends to be small when¬†$ùëå$ is large. <br>

$$\text{Cor}(ùëã,ùëå)=\rho=\dfrac{\text{Cov}(ùëã,ùëå)}{\sigma_X \sigma_Y}, \qquad -1\leq\rho\leq 1$$
$\rho>0$:	large values of¬†$ùëã$ tend to happen with large values of¬†$ùëå$. <br>
$\rho<0$:	$ùëã$ tends to be small when¬†$ùëå$ is large. <br>

```python
temperature = [24.56, 26.00, 27.79, 20.59, 11.50, 14.35, 13.70]
icecream = [534.79,625.19,660.63,487.70,316.24,367.94,308.89]
hotpot = [385.21, 343.62, 364.02, 388.78, 781.34, 483.24, 420.41]
gold = [1932.33, 1913.92, 1921.48, 1980.02, 1909.81, 1944.26, 1908.27]
```

```python
data = np.transpose([temperature, icecream, hotpot, gold])
data = pd.DataFrame(data, columns=['temperature','icecream','hotpot', 'gold'])
print(data)
```

```python
data_cov = np.cov([temperature, icecream, hotpot, gold])

# print
name = ["temperature", "icecream", "hotpot", "gold"]
print(" "*15, name)
for _i, _j in zip(name, data_cov):
    print(f"{_i:<15}", _j)
```

```python
data_corr = np.corrcoef([temperature, icecream, hotpot, gold])

# print
name = ["temperature", "icecream", "hotpot", "gold"]
print(" "*15, name)
for _i, _j in zip(name, data_corr):
    print(f"{_i:<15}", _j)
```
